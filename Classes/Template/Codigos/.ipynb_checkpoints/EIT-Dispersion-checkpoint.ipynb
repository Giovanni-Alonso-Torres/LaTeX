```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi as π
```

```python
δp = np.linspace(-120, 120, 10000)
Γe = 2*π*6.06 # En MHz
Γr = 2*π*0.1  # En MHz
γg = 0.15     # En MHz

def χ(Ωc):
    return (δp+1j*(Γr/2+2*γg))/((Γr/2+2*γg-1j*δp)*(Γe/2+γg/2-1j*δp)+Ωc**2/4)
```

```python
plt.figure(figsize=(8,6))
plt.plot(δp, χ(2*π*1).real, color='dodgerblue', label='a')
plt.plot(δp, χ(0).real, color='orangered')
plt.xticks([0],[0])
plt.yticks([0],[0])
plt.legend(fontsize=16)
plt.show()
```
